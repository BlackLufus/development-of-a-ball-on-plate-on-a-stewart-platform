## Import necessary moduls

```python
import numpy as np
import matplotlib.pyplot as plt
```

## P - Regler

```python
# Parameters
Kp = 4.0

x = 0.2 # Current position in meter
v = 0.0 # Current velocity in m/s
x_target = 0.0 # Target position in meter
g = 9.81 # Gravity in m/s^2
dt = 0.1 # 100 ms in seconds
max_angle = np.radians(4)  # rad
steps = 100

# Storage
positions = []
angles = []

for _ in range(steps):
    # Errorcalculation
    e = x_target - x

    # P-Controller
    u = Kp * e

    # P - Controller
    u = np.clip(u, -max_angle, max_angle)

    # Physics
    a = g * np.sin(u)
    v += a * dt
    x += v * dt

    # Store values
    positions.append(x)
    angles.append(np.degrees(u))

# Plot
fig, axis = plt.subplots(1, 2, figsize=(10, 4))
axis[0].plot(positions)
axis[0].set_title("Ballposition (x)")
axis[0].set_xlabel("Time (s)")
axis[0].set_ylabel("Position (m)")

axis[1].plot(angles)
axis[1].set_title("Roll Angle")
axis[1].set_xlabel("Time (s)")
axis[1].set_ylabel("Angle ()")

plt.show()
```

## PID - Regler

```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider

def simulate_pid(Kp=4.0, Ki=2.0, Kd=1.0):
    # Parameters
    x = 0.2 # Current position in meter
    v = 0.0 # Current velocity in m/s
    x_target = 0.0 # Target position in meter
    g = 9.81 # Gravity in m/s^2
    dt = 0.1 # 100 ms in seconds
    max_angle = np.radians(4)  # rad
    steps = 100

    # State
    integral = 0.0
    prev_error = 0.0

    # Storage
    positions = []
    angles = []

    for _ in range(steps):
        e = x_target - x
        integral += e * dt
        derivative = (e - prev_error) / dt
        prev_error = e

        # PID - Controller
        u = Kp * e + Ki * integral + Kd * derivative
        u = np.clip(u, -max_angle, max_angle)

        # Physics
        a = g * np.sin(u)
        v += a * dt
        x += v * dt

        positions.append(x)
        angles.append(np.degrees(u))

    # Plot
    fig, axis = plt.subplots(1, 2, figsize=(10, 4))
    axis[0].plot(positions)
    axis[0].set_title("Ballposition (x)")
    axis[0].set_xlabel("Time (s)")
    axis[0].set_ylabel("Position (m)")

    axis[1].plot(angles)
    axis[1].set_title("Roll Angle")
    axis[1].set_xlabel("Time (s)")
    axis[1].set_ylabel("Angle ()")

    plt.show()

# Interaktive Slider
interact(simulate_pid,
         Kp=FloatSlider(min=0, max=10, step=0.1, value=4.0, description='Kp'),
         Ki=FloatSlider(min=0, max=10, step=0.1, value=2.0, description='Ki'),
         Kd=FloatSlider(min=0, max=10, step=0.1, value=1.0, description='Kd'))
```
